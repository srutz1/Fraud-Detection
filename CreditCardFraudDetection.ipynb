Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a Pandas Dataframe
data=pd.read_csv('/content/creditcard.csv')
```

```python
#checking the number of missing values in each column
data.isnull().sum()
```

```python
#removing NULL vallue row from the dataset
data = data.dropna()
```

```python
#distribution of legit transactions & fraudelent transaction
data['Class'].value_counts()
```

```python
#separating the data for analysis
legit= data[data.Class==0]
fraud= data[data.Class==1]
```

UnderSampling

*   Build a sample dataset containing similar distribution of normal transactions and Fradulent Transactions
*   Number of Fraudulent Transactions -->88



```python
legit_sample=legit.sample(n=88)
#concatenating two dataframes
#axis =0 means row
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
#this is for checking wheter it is a good data set or not
#if the difference of means is more it is not good sample
#if less it is good sample
```

```python
#Split the data into Features And Targets
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
print (X)
```

```python
print(Y)
```

Split the data into Training Data and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
#test_size=0.2 show that 20 persentage goes to test data and 80 to train data
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression(solver='liblinear', C=0.01)
#solver='liblinear' beacause of small dataset
```

```python
#training the Logistic Regression Model with the training Data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training Data:', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Test Data:', test_data_accuracy)
```
